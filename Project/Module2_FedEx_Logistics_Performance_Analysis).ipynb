````python
# Import the dataset from Google Drive into a pandas DataFrame
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AlmaBetter/Final-Project-Files-Module-2/Capstone Project - 2 - FedEx Logistics Performance Analysis /FedEx Logistics Performance Analysis Dataset/SCMS_Delivery_History_Dataset.csv')

# Display concise summary of the DataFrame
# Includes column names, data types, and non-null values
df.info()

# Generate statistical summary of numerical columns
# Shows count, mean, std, min, max, and quartiles
df.describe()

# Get the shape of the dataset (rows, columns)
df.shape

# Display the first 5 rows of the dataset
df.head()

# Display the last 5 rows of the dataset
df.tail()

# Show all column names in the dataset
df.columns

# Check number of missing (null) values in each column
df.isnull().sum()

# Count number of unique values in each column
# dropna=False ensures NaN values are also counted
df.nunique(dropna=False)

# Count total duplicate rows in the dataset
df.duplicated().sum()

# Remove duplicate rows from the dataset
# inplace=True modifies the DataFrame directly
df.drop_duplicates(inplace=True)
````

## 3. ***Data Wrangling***


## Data Cleaning

Steps performed:

- Remove duplicates
- Convert date columns
- Fill missing values
- Standardize country names

## Feature Engineering

- New variables are created to improve analysis.

### Created Columns
- Lead Time Days
- Delay Days
- OnTime Delivery


````python
# Convert date columns
date_columns = ['Scheduled Delivery Date', 'Delivered to Client Date']
df[date_columns] = df[date_columns].apply(pd.to_datetime, errors='coerce')

# Handle missing values
df['Shipment Mode'] = df['Shipment Mode'].fillna(df['Shipment Mode'].mode()[0])
df['Line Item Insurance (USD)'] = df['Line Item Insurance (USD)'].fillna(0)

# Clean country names
df['Country'] = df['Country'].str.strip().str.lower()

# Convert numeric columns
numeric_columns = ['Weight (Kilograms)', 'Freight Cost (USD)']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Create new features
df['Lead Time Days'] = (df['Delivered to Client Date'] - df['Scheduled Delivery Date']).dt.days
df['Delay Days'] = df['Lead Time Days']
df['OnTime'] = (df['Delivered to Client Date'] <= df['Scheduled Delivery Date']).astype(int)

# Convert 'Weight (Kilograms)' and 'Freight Cost (USD)' to numeric, coercing errors
#df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')
#df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')

# Display info again to confirm data types
df[['Weight (Kilograms)', 'Freight Cost (USD)']].info()
````

**Chart 7 — Average Lead Time by Year**

````python
# Create 'Year' column
df['Year'] = df['Delivered to Client Date'].dt.year

# Calculate average Lead Time per year
yearly_lead_time = df.groupby('Year')['Lead Time Days'].mean().reset_index()

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(
    yearly_lead_time['Year'].astype(str),
    yearly_lead_time['Lead Time Days'],
    marker='o',
    linestyle='-',
    color='#ff7f0e',
    label='Avg Lead Time'
)
plt.title('Chart 7 - Average Lead Time Per Year')
plt.xlabel('Year')
plt.ylabel('Lead Time (Days)')
plt.grid(True, linestyle='--', alpha=0.9)
plt.legend()
plt.tight_layout()
plt.show()
````

**Chart 9 — Top 10 Countries by Delayed Shipments**

````python
# Map boolean to readable labels
df['Delayed'] = df['Lead Time Days'] > 0
df['DeliveryStatus'] = df['Delayed'].map({False:'Delayed', True:'On-Time'})

# Aggregate counts by Country and Delivery Status
agg_data = df.groupby(['Country', 'DeliveryStatus']).size().unstack(fill_value=0)

# Sort countries by number of delayed shipments and take top 10
top10_df = agg_data.sort_values('Delayed', ascending=False).head(10)

# Define colors: Delayed = red, On-Time = green
colors = ['#d62728', '#2ca02c']

# Plot side-by-side bars
top10_df.plot(
    kind='bar',
    figsize=(12,6),
    color=colors,
    width=0.6
)

plt.title('Chart 9 - Top 10 Countries by Delayed Shipments')
plt.xlabel('Country')
plt.ylabel('Number of Shipments')
plt.xticks(rotation=0)
plt.legend(title='Delivery Status')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

**Why did you pick this chart?**

A grouped bar chart (side-by-side bars) is ideal for comparing categories across multiple groups. Here, it allows us to easily compare On-Time vs Delayed shipments for the top 10 countries.

**Insights**

Shipment performance varies significantly by country.

Some countries have a higher proportion of on-time deliveries, while others experience more delays.

Highlights operational strengths and areas needing improvement across regions.

**Positive Impact**

Identifies countries with strong delivery performance, which can serve as benchmarks.

**Chart 10 — Correlation Heatmap**
